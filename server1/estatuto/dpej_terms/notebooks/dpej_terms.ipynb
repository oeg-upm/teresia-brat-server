```python
import os
import pandas as pd

# Ruta al directorio donde están los archivos `.ann`
path = "/Users/jpardo/Desktop/Proyectos/TeresIA/Recursos/server/dpej_terms"

# Obtener la lista de archivos en el directorio
archivos = os.listdir(path)

# Lista para almacenar los datos de cada término
data = []

# Procesar cada archivo `.ann`
for archivo in archivos:
    if archivo.endswith('.ann'):
        with open(os.path.join(path, archivo), 'r') as file:
            # Leer cada línea del archivo y extraer información
            for line in file:
                # Separar en columnas por tabulaciones
                cols = line.strip().split('\t')
                if len(cols) >= 3:
                    term_id = cols[0]
                    term_info = cols[1].split()  # Ej: ['term', '1122', '1130']
                    term_text = cols[2]
                    
                    # Estructurar los datos
                    data.append({
                        "Documento": archivo,
                        "ID Término": term_id,
                        "Tipo": term_info[0],
                        "Inicio": int(term_info[1]),
                        "Fin": int(term_info[2]),
                        "Texto": term_text
                    })

# Crear un DataFrame de pandas con los datos
df = pd.DataFrame(data)

# Mostrar la tabla
df

```

```python
# NOTA: ESTA FUNCION ESTA MAL

def buscar_terminos_por_span(texto, tabla_terminos):
    """
    Busca términos en el texto usando los spans de la tabla de términos y verifica si coinciden.

    :param texto: str - Texto donde se buscan los términos.
    :param tabla_terminos: pd.DataFrame - DataFrame con las columnas "Inicio", "Fin" y "Texto" de términos anotados.
    :return: List[Dict] - Lista de diccionarios con los términos encontrados y sus spans.
    """
    resultados = []

    for _, fila in tabla_terminos.iterrows():
        inicio = fila['Inicio']
        fin = fila['Fin']
        texto_anotado = fila['Texto']

        # Extraer el texto correspondiente al span en el texto
        texto_span = texto[inicio:fin]

        # Verificar si el texto en el span coincide con el texto anotado
        if texto_span == texto_anotado:
            resultados.append({
                "Inicio": inicio,
                "Fin": fin,
                "Texto": texto_anotado
            })
        else:
            print(f"El texto encontrado en ({inicio}, {fin}) es: {texto_span}")

    return resultados
```

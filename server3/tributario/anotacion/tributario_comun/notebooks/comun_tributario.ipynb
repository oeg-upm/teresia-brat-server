```python
# ---------- COMBINAR .ann REMAPEANDO IDs (T y referencias) ----------

import os
import re
from typing import Dict, List

ID_T_RE = re.compile(r'\bT\d+\b')   # para sustituir en atributos y argumentos

def procesar_anotaciones(general, selected_file, base, suffixes, output_file):
    """
    Combina un .ann (selected_file) de múltiples carpetas indicadas en 'general',
    reenumerando entidades (T) y relaciones (R) de forma global y consistente.

    - Remapea TODAS las referencias a T (en R, A, M, N, #) al nuevo ID.
    - Cambia 'concept' -> 'Term-<suffix>' según tu regla.
    - Añade sufijo a los tipos de entidad y de relación: <label>-<suffix>.
    - Soporta spans múltiples (e.g., "start end;start end").
    - Escribe entidades primero, luego atributos, luego relaciones.

    Params
    ------
    general : list[dict]   Cada item: {'path': str, 'archivos': [str, ...]}
    selected_file : str    Nombre de .ann a buscar en cada 'path'
    base : str             Carpeta base (raíz)
    suffixes : dict        Mapa: path -> sufijo (p.ej. {'carpeta1': 'est', ...})
    output_file : str      Ruta de salida para el .ann combinado
    """
    entities = []     # líneas T remapeadas
    attributes = []   # líneas A/M/N/# remapeadas
    relations = []    # líneas R remapeadas

    term_counter = 1
    relation_counter = 1
    attr_counter = 1  # si hay atributos sin ID propio (A1), se puede reenumerar

    # Recorremos cada origen
    for item in general:
        path = item['path']
        ann_files = [f for f in item['archivos'] if f == selected_file]
        suffix = suffixes.get(path, "")

        for ann_file in ann_files:
            full_path = os.path.join(base, path, ann_file)
            lines = carga_ann(full_path)  # <- tu función que devuelve lista de líneas .ann

            # Mapa por archivo: T_old -> T_new
            tmap: Dict[str, str] = {}

            # 1) Primero pasamos por todas las entidades (T...) para crear el mapa
            for line in lines:
                if not line or line[0] != 'T':
                    continue

                # BRAT: T<ID>\t<TYPE> <SPAN(s)>\t<TEXT>
                chunks = line.rstrip('\n').split('\t')
                if len(chunks) < 3:
                    continue  # línea T mal formada

                old_tid = chunks[0]          # p.ej. "T12"
                type_and_spans = chunks[1]   # p.ej. "concept 10 20" o "concept 10 20;30 40"
                text = chunks[2]             # superficie

                # separar label y spans
                parts = type_and_spans.split(' ', 1)
                if len(parts) != 2:
                    continue
                raw_label, span_str = parts[0], parts[1]

                # normalizar label (concept -> Term-<suffix>, resto añade sufijo)
                if raw_label.lower() == "concept":
                    label = f"Term-{suffix}" if suffix else "Term"
                else:
                    label_norm = raw_label  # puedes sanear si necesitas
                    label = f"{label_norm}-{suffix}" if suffix else label_norm

                # asignar nuevo ID
                new_tid = f"T{term_counter}"
                term_counter += 1
                tmap[old_tid] = new_tid

                # reconstruir línea T con el nuevo tipo (y mismo span/text)
                entities.append(f"{new_tid}\t{label} {span_str}\t{text}")

            # 2) Segundo pase: atributos y normalizaciones (A/M/N/#)
            for line in lines:
                if not line or line[0] in ('T', 'R'):
                    continue

                # Ejemplos:
                # A1\tSomeAttr T3 1
                # M1\tReference T4 ID Wikipedia:12345
                # N1\tReference T7 Wikipedia "xxxx"
                # #1\tAnnotatorNotes T3 texto...
                # Necesitamos remapear los T dentro
                # Intentamos preservar el ID del atributo si ya existe; si no, reenumeramos.
                l = line.rstrip('\n')
                parts_tab = l.split('\t', 1)
                if len(parts_tab) == 2 and parts_tab[0] and parts_tab[0][0] in ('A', 'M', 'N', '#'):
                    head_id = parts_tab[0]  # A1 / M1 / N1 / #1 ...
                    rest = parts_tab[1]

                    # reemplazar Txx por su nuevo ID con el mapa
                    def _sub_tids(m):
                        old = m.group(0)
                        return tmap.get(old, old)

                    rest_remap = ID_T_RE.sub(_sub_tids, rest)
                    # conservamos ID original del atributo si no choca
                    attributes.append(f"{head_id}\t{rest_remap}")
                else:
                    # línea rara/no estándar: remapea Txx por si acaso
                    def _sub_tids(m):
                        old = m.group(0)
                        return tmap.get(old, old)
                    attributes.append(ID_T_RE.sub(_sub_tids, l))

            # 3) Tercer pase: relaciones (R...)
            for line in lines:
                if not line or line[0] != 'R':
                    continue

                # BRAT: R<ID>\t<REL> Arg1:Tn Arg2:Tm [Arg3:Tx ...]
                l = line.rstrip('\n')
                parts_tab = l.split('\t')
                if len(parts_tab) < 2:
                    continue

                # ignoramos el viejo R-id; creamos uno nuevo global
                # parts_tab[1] = "<REL> Arg1:Tn Arg2:Tm ..."
                rel_and_args = parts_tab[1].split(' ', 1)
                if len(rel_and_args) != 2:
                    continue
                raw_rel = rel_and_args[0]
                args = rel_and_args[1]

                # añadir sufijo al label de relación
                rel_label = f"{raw_rel}-{suffix}" if suffix else raw_rel

                # remapear todos los Txx en los argumentos
                def _sub_tids(m):
                    old = m.group(0)
                    return tmap.get(old, old)

                args_remap = ID_T_RE.sub(_sub_tids, args)

                # construir nueva línea R
                new_rid = f"R{relation_counter}"
                relation_counter += 1
                relations.append(f"{new_rid}\t{rel_label} {args_remap}")

    # 4) Escribir salida: entidades, atributos, relaciones
    with open(output_file, 'w', encoding='utf-8') as out:
        for t in entities:
            out.write(t + '\n')
        for a in attributes:
            out.write(a + '\n')
        for r in relations:
            out.write(r + '\n')

    print(f"[OK] Guardado {output_file}  (T={len(entities)}  A={len(attributes)}  R={len(relations)})")
```

```python
def preparar_rutas(df, base_path, output_path):
    """
    Para cada fila del DataFrame, construye dos columnas:
    - 'selected_file': base_path + path + ann_file
    - 'output_file':   output_path + ann_file
    """
    # Crear la columna 'selected_file'
    df["selected_file"] = df.apply(
        lambda row: os.path.join(base_path, row["path"], row["ann_file"]), 
        axis=1
    )

    # Crear la columna 'output_file'
    df["output_file"] = df.apply(
        lambda row: os.path.join(output_path, row["ann_file"]), 
        axis=1
    )

    return df
```

```python
from pathlib import Path
from collections import defaultdict
import re
from typing import Dict, List, Set, Tuple, Iterable, Optional

ANN_LINE_SPLIT = re.compile(r"\t")
```

```python
def _read_ann_lines(path: Path) -> List[str]:
    return [ln.rstrip("\n") for ln in path.read_text(encoding="utf-8", errors="replace").splitlines() if ln.strip()]

def _parse_entity(line: str):
    # T12\tLabel start end(;start end ...)\ttext...
    tid, mid, text = ANN_LINE_SPLIT.split(line, maxsplit=2)
    label, spans = mid.split(" ", 1)
    return tid, label, spans, text

def _parse_relation(line: str):
    # R1\tRel Arg1:T2 Arg2:T3 (posibles args extra)
    rid, rest = ANN_LINE_SPLIT.split(line, maxsplit=1)
    parts = rest.split()
    rel = parts[0]
    args = {}
    for tok in parts[1:]:
        if ":" in tok:
            k, v = tok.split(":", 1)
            args[k] = v
    return rid, rel, args  # args['Arg1']='T2', args['Arg2']='T3', ...

def _parse_attribute(line: str):
    # A1\tAttrName T3 [Value]
    # M/N/# también pueden referir a T/R/E; aquí nos importa el target y el posible valor al final
    # Devolvemos (head_id, attr_name, target_id, value_or_None)
    head, rest = ANN_LINE_SPLIT.split(line, maxsplit=1)
    parts = rest.split()
    if not parts:
        return head, None, None, None
    attr_name = parts[0]
    # target suele ser el segundo token y tiene forma Txx/Rxx/Exx
    target_id = None
    value = None
    for tok in parts[1:]:
        if tok[:1] in ("T", "R", "E") and tok[1:].isdigit():
            target_id = tok
        else:
            # el primer token que no sea target lo interpretamos como valor (si existe)
            if value is None:
                value = tok
    return head, attr_name, target_id, value

def collect_from_anns(
    ann_paths: Iterable[Path]
):
    """
    Recorre .ann y devuelve:
      - entity_types: set de labels T
      - rel2pairs:    dict rel -> set de (Arg1_type, Arg2_type) observados
      - attr_defs:    dict attr -> {'targets_entity_types': set, 'targets_relation': bool, 'targets_event': bool, 'values': set}
    """
    entity_types: Set[str] = set()
    rel2pairs: Dict[str, Set[Tuple[str, str]]] = defaultdict(set)
    attr_defs = defaultdict(lambda: {
        "targets_entity_types": set(),  # tipos concretos T
        "targets_relation": False,      # si alguna vez apuntó a R
        "targets_event": False,         # si alguna vez apuntó a E
        "values": set()                 # valores literales observados
    })

    for ann_path in ann_paths:
        if not ann_path.exists() or ann_path.suffix != ".ann":
            continue
        lines = _read_ann_lines(ann_path)

        # Mapa por documento: T-id -> label
        tlabel: Dict[str, str] = {}

        # Primer pase: entidades
        for ln in lines:
            if ln.startswith("T"):
                tid, label, spans, text = _parse_entity(ln)
                tlabel[tid] = label
                entity_types.add(label)

        # Segundo pase: relaciones y atributos
        for ln in lines:
            tag = ln[:1]
            if tag == "R":
                _, rel, args = _parse_relation(ln)
                a1 = args.get("Arg1")
                a2 = args.get("Arg2")
                if a1 and a2 and a1 in tlabel and a2 in tlabel:
                    rel2pairs[rel].add((tlabel[a1], tlabel[a2]))
            elif tag in ("A", "M", "N", "#"):
                _, attr, target, value = _parse_attribute(ln)
                if not attr or not target:
                    continue
                if target.startswith("T"):
                    # atributo sobre entidad concreta
                    # si conocemos el tipo exacto, lo añadimos
                    if target in tlabel:
                        attr_defs[attr]["targets_entity_types"].add(tlabel[target])
                elif target.startswith("R"):
                    attr_defs[attr]["targets_relation"] = True
                elif target.startswith("E"):
                    attr_defs[attr]["targets_event"] = True
                if value:
                    attr_defs[attr]["values"].add(value)

    return entity_types, rel2pairs, attr_defs
```

```python

def write_annotation_conf(
    out_path: Path,
    entity_types: Set[str],
    rel2pairs: Dict[str, Set[Tuple[str, str]]],
    attr_defs
):
    """
    Escribe annotation.conf con secciones [entities], [relations], [events], [attributes],
    ajustado estrictamente a lo observado en los .ann.
    """
    lines: List[str] = []

    # --- entities ---
    lines.append("[entities]")
    for et in sorted(entity_types):
        lines.append(et)
    lines.append("")

    # --- relations ---
    lines.append("[relations]")
    # Para no crear combinaciones nuevas: una línea por PAR observado
    for rel in sorted(rel2pairs.keys()):
        pairs = sorted(rel2pairs[rel])
        for (t1, t2) in pairs:
            lines.append(f"{rel} Arg1:{t1}, Arg2:{t2}")
    lines.append("")

    # --- events ---
    # Si no tienes eventos E en tus .ann, lo dejamos vacío (BRAT exige la sección).
    lines.append("[events]")
    lines.append("")

    # --- attributes ---
    lines.append("[attributes]")
    # Atributos booleanos o multivaluados; Arg: a tipos concretos o comodines BRAT.
    # Doc oficial: Value:V1|V2|... ; Arg:<EVENT>/<RELATION>/lista de tipos.  :contentReference[oaicite:1]{index=1}
    for attr in sorted(attr_defs.keys()):
        info = attr_defs[attr]
        targets: List[str] = []
        ent_types = sorted(info["targets_entity_types"])
        if ent_types:
            targets.append("|".join(ent_types))
        if info["targets_relation"]:
            targets.append("<RELATION>")
        if info["targets_event"]:
            targets.append("<EVENT>")

        if not targets:
            # fallback prudente: si no pudimos inferir target, apuntamos a entidades
            targets = ["<ENTITY>"]

        arg_spec = "Arg:" + "|".join(targets)

        values = sorted(info["values"])
        if values:
            # atributo multivaluado
            val_spec = "Value:" + "|".join(values)
            lines.append(f"{attr} {arg_spec}, {val_spec}")
        else:
            # atributo booleano
            lines.append(f"{attr} {arg_spec}")
    lines.append("")

    out_path.write_text("\n".join(lines), encoding="utf-8")
    return out_path
```

## visual

```python
from pathlib import Path

def write_visual_conf(
    out_path: Path,
    entity_types,            # set[str]
    relation_types,          # set[str]
    attribute_names          # set[str]
):
    """
    Genera visual.conf con:
      - [labels] para todas las entidades, relaciones y atributos presentes en los .ann
      - [drawing] vacío (usa estilos por defecto de BRAT)

    No inventa etiquetas ni colores: identidad pura.
    """
    lines = []

    # --- [labels] ---
    lines.append("[labels]")
    # Entidades
    for et in sorted(entity_types):
        lines.append(f"{et} | {et}")
    # Relaciones
    for rel in sorted(relation_types):
        lines.append(f"{rel} | {rel}")
    # Atributos (opcionales; aparecen tal cual)
    for attr in sorted(attribute_names):
        lines.append(f"{attr} | {attr}")
    lines.append("")

    # --- [drawing] ---
    # Sección requerida; dejamos vacío para no introducir estilos nuevos.
    lines.append("[drawing]")
    lines.append("")

    out_path.write_text("\n".join(lines), encoding="utf-8")
    return out_path
```

## tools

```python
# =========================
# tools.conf a partir de .ann
# =========================
from pathlib import Path
from typing import Iterable, List, Set, Tuple, Dict
import re

# # --- Reutiliza _read_ann_lines si ya lo tienes; si no, te lo dejo aquí:
# def _read_ann_lines(path: Path) -> List[str]:
#     return [ln.rstrip("\n") for ln in path.read_text(encoding="utf-8", errors="replace").splitlines() if ln.strip()]

# --------- 1) Recolector de recursos de normalización (RID en 'N' lines)
# Formato N típico (doc oficial):
# N1\tReference T1 Wikipedia:534366 Barack Obama
# => RID = 'Wikipedia', EID = '534366'
N_LINE_RE = re.compile(r"^N\d+\tReference\s+(T\d+|R\d+|E\d+)\s+([A-Za-z0-9_\-]+):(\S+)\b")

def collect_normalization_resources(ann_paths: Iterable[Path]) -> Set[str]:
    """
    Devuelve el conjunto de RID (nombres de recursos de normalización) observados
    en los .ann (líneas que empiezan por 'N' con tipo 'Reference').
    """
    rids: Set[str] = set()
    for p in ann_paths:
        if not p.exists() or p.suffix != ".ann":
            continue
        for ln in _read_ann_lines(p):
            if not ln or ln[0] != "N":
                continue
            m = N_LINE_RE.match(ln)
            if m:
                rid = m.group(2)  # e.g., 'Wikipedia'
                if rid:
                    rids.add(rid)
    return rids
```

```python

# --------- 2) Escritura de tools.conf
# Mapa opcional de recursos "conocidos" -> URLs canónicas (no añade recursos nuevos)
KNOWN_NORM_URLS: Dict[str, Tuple[str, str]] = {
    # DBNAME: (HOMEURL, ENTRYURL)
    "Wikipedia": ("http://en.wikipedia.org", "http://en.wikipedia.org/?curid=%s"),
    "Wiki":      ("http://en.wikipedia.org", "http://en.wikipedia.org/?curid=%s"),
    "Wikidata":  ("https://www.wikidata.org", "https://www.wikidata.org/wiki/%s"),
    "DBpedia":   ("http://dbpedia.org", "http://dbpedia.org/resource/%s"),
    # añade aquí SOLO si aparece en tus .ann y conoces la URL oficial
}

def write_tools_conf(
    out_path: Path,
    norm_resources: Set[str],
    write_empty_sections: bool = True
) -> Path:
    """
    Genera tools.conf con las secciones estándar y SOLO los recursos de normalización
    observados en .ann (norm_resources). No inventa recursos; para conocidos añade URLs.
    """
    lines: List[str] = []

    # [options] — lo dejamos vacío para no forzar tokenizador/validador
    if write_empty_sections:
        lines.append("[options]")
        lines.append("")

    # [search] — sin entradas nuevas (no generamos proveedores de búsqueda)
    if write_empty_sections:
        lines.append("[search]")
        lines.append("")

    # [normalization] — aquí reflejamos EXACTAMENTE los RID observados
    lines.append("[normalization]")
    for rid in sorted(norm_resources):
        if rid in KNOWN_NORM_URLS:
            home, entry = KNOWN_NORM_URLS[rid]
            # DB: (ruta) es opcional; URLBASE también opcional, pero útil si la sabemos.
            lines.append(f"{rid} <URL>:{home}, <URLBASE>:{entry}")
        else:
            # Entrada mínima sin adivinar; válida, BRAT permite tools.conf vacío. 
            # Si tienes el DB creado con ese nombre, lo reconocerá en su ubicación por defecto.
            lines.append(f"{rid}")
    lines.append("")

    # [annotators] — no declaramos servicios automáticos (no generamos nada nuevo)
    if write_empty_sections:
        lines.append("[annotators]")
        lines.append("")

    # [disambiguators] — igual, vacío
    if write_empty_sections:
        lines.append("[disambiguators]")
        lines.append("")

    out_path.write_text("\n".join(lines), encoding="utf-8")
    return out_path
```
